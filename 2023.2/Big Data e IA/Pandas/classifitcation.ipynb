```python
import matplotlib.pyplot  as plt
from sklearn import datasets
from sklearn import tree
```

```python
iris = datasets.load_iris()
```

```python
data = iris.data
target = iris.target
```

```python
print('Data')
print(data[:5])
```

```python
print('Target')
print(target[:5])
```

```python
X, y = iris.data, iris.target
fig = plt.figure(figsize=(30, 25))
classfier = tree.DecisionTreeClassifier()
classfier.fit(X, y)
tree.plot_tree(classfier, feature_names=iris.feature_names, class_names=iris.target_names.tolist(), filled=True)
plt.savefig('dt.png')
plt.show()
```
